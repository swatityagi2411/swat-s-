```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('flipkart_products.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
x=df['title'].values
y=df['price'].values
```

```python
numer=0
denom=0
n=len(x)
x_mean=np.mean(x)
y_mean=np.mean(y)
for i in range(n):
  numer+=(x[i]-x_mean)*(y[i]-y_mean)
  denom=(x[i]-x_mean)**2

slope=numer/denom
intercept=y_mean-slope*x_mean
```

```python
slope,intercept
```

```python
prediction=[]
for i in range(n):
  y_pred=slope*x[i]+intercept
  prediction.append(int(y_pred))
```

```python
prediction[:5]
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y)
plt.plot(x,prediction,'red')
plt.show()
```

```python
err=0
for i in range(n):
  err+=prediction[i]-y[i]**2
mse=err/n
```

```python
print("mean squared error=",mse)
np.sqrt(mse)
```
