```python
import matplotlib.pyplot as plt
names=["Ram","Jaes","Karry","Roun","Tom"]
salary=[20000,500,10000,18000,270000]
prevsalary=[10000,200,40000,28567,83829]
data={
    "names":names,
    "salary":salary,
    "prevsalary":prevsalary
}
plt.style.use('ggplot')
plt.figure(figsize=(10,4))
#plt.bar(data["names"],data["salary"],color='green')
plt.barh(data["names"],data["salary"],color='green')
plt.title('Salary Comparison')
#plt.xlabel('Names')
#plt.ylabel('Salary')
plt.ylabel('Names')
plt.xlabel('Salary')
plt.show()
```

```python
plt.figure(figsize=(10,6))
#plt.bar(data["names"],data["salary"],color='green')
plt.plot(data["names"],data["salary"],color='green')
plt.title('Salary Comparison')
```

```python
plt.figure(figsize=(10,4))
#plt.bar(data["names"],data["salary"],color='green')
plt.scatter(data["names"],data["salary"],color='blue',marker='*')
plt.title('Salary Comparison')
```

```python
plt.figure(figsize=(10,4))
#plt.bar(data["names"],data["salary"],color='green')
plt.plot(data["names"],data["salary"],marker='*',label='current')
plt.plot(data["names"],data["prevsalary"],marker='*',label='prev')
plt.title('Salary Comparison')
plt.legend()
```
