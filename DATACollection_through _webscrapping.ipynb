**Collecting and storing numerous pages data in excel format from flipkart**

```python
import bs4
import urllib.request as url
import pandas as pd
data={"title":[],
      "price":[]}
for i in range(1,5):
  print("page",i)
  path=f"https://www.flipkart.com/search?q=i+phone+14pro+max&page={i}"
  response=url.urlopen(path)
  page=bs4.BeautifulSoup(response)
  titlelist=page.find_all('div',{'class':'_4rR01T'})
  pricelist=page.find_all('div',{'class':'_30jeq3 _1_WHN1'})
  for j in range(len(titlelist)):
    data["title"].append(titlelist[j].text)
    data["price"].append(pricelist[j].text)
   
```

```python
df=pd.DataFrame(data)
```
